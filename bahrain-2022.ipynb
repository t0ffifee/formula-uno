```python
import fastf1
import fastf1.plotting
fastf1.plotting.setup_mpl()
fastf1.Cache.enable_cache('cache')

from functions import *

year = 2022
circuit = 'bahrain'
event = 'R'

session = fastf1.get_session(year, circuit, event)
session.load()

lec_d, lec_v = get_fastests_d_v('LEC', session)
ver_d, ver_v = get_fastests_d_v('VER', session)
ham_d, ham_v = get_fastests_d_v('HAM', session)
ric_d, ric_v = get_fastests_d_v('RIC', session)
```

```python
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')

rbr_color = fastf1.plotting.team_color('RBR')
fer_color = fastf1.plotting.team_color('FER')
mer_color = fastf1.plotting.team_color('MER')
mcl_color = fastf1.plotting.team_color('MCL')

fig, ax = plt.subplots()
ax.plot(ver_d, ver_v, color=rbr_color, label='VER')
# ax.plot(lec_d, lec_v, color=fer_color, label='LEC')
# ax.plot(ham_d, ham_v, color=mer_color, label='HAM')
ax.plot(ric_d, ric_v, color=mcl_color, label='RIC')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Race")

fig.set_dpi(150)
ax.set(xlabel='Distance', ylabel='Speed')

plt.show()
```
