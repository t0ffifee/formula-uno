```python
import fastf1
```

```python
session = fastf1.get_session(2024, 1, 'R')
session.load()
```

```python
session.results
```

```python
f1_points = [25, 18, 15, 12, 10, 8, 6, 4, 2, 1]
driver_and_team = list(reversed(list(zip(session.results['FullName'], session.results['TeamName']))))

toilet_bowl_ranking = list(zip(driver_and_team, f1_points))
```

```python
print("\t===== Toilet Bowl Championships =====\n")
for (driver, team), points in toilet_bowl_ranking:
  print(driver + "\t\t" + team + "\t\t\t" + str(points))
```
